```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head() 
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='magma')
```
