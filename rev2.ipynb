```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'indigo')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.2,0.5,.2,.2])
axes.plot(x,y)
axes1.plot(x,y)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z,'indigo')
ax2.plot(x,y,'indigo')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.plot(x,y)
```

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))

axes[0].plot(x,y,'blue',ls='--',lw=3)
axes[1].plot(x,z,'red',ls='-',lw=4)
axes[1].text(0.15, 0.2, r"$y=x^2$", fontsize=15, color="blue")

plt.tight_layout()
```
