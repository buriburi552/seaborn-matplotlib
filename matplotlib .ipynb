```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # rows, columns, plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left,bottom,width,height

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("first plot")
axes2.set_title("second plot")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes is an array which means we can itterate through this and call indexes

# for curr_axes in axes:
#     curr_axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("first plot")

axes[1].plot(y,x)
axes[1].set_title("Second plot")
plt.tight_layout() # to remove overlapping
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1]) 
                                
ax.plot(x,y,color='#FF8C00',linewidth=2,alpha=0.5,linestyle='--')

#go to notebook

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
