```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(tips['total_bill'],kde=True,bins=30)
sns.set_theme(style='darkgrid',palette='coolwarm')
```

```python
sns.jointplot(x = tips['total_bill'],y = tips['tip'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm',height=3)
```

```python
sns.rugplot(data=tips,x='total_bill')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data = tips,palette='magma')
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data = tips,hue='smoker',palette='coolwarm')
```
